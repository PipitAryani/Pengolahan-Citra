```python
# Install library yang dibutuhkan
!pip install deepface opencv-python matplotlib

# Import library
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt

# Daftar 10 gambar wajah (ganti path sesuai file yang diunggah)
# Contoh: setelah upload ke /content/
image_paths = [
    "/content/wajah1.jpg",
    "/content/wajah2.jpg",
    "/content/wajah3.jpg",
    "/content/wajah4.jpg",
    "/content/wajah5.jpg",
    "/content/wajah6.jpg",
    "/content/wajah7.jpg",
    "/content/wajah8.jpg",
    "/content/wajah9.jpg",
    "/content/wajah10.jpg"
]

# Daftar nama untuk tiap gambar (sesuaikan)
names = [
    "Andi", "Budi", "Citra", "Dewi", "Eka",
    "Farhan", "Gita", "Hendra", "Intan", "Joko"
]

# Deteksi wajah dan prediksi usia
results = []

for i, path in enumerate(image_paths):
    try:
        # Analisis wajah dengan DeepFace
        analysis = DeepFace.analyze(img_path=path, actions=['age'], enforce_detection=False)
        age = analysis[0]['age']

        # Simpan hasil
        results.append((names[i], age, path))

        # Tampilkan hasil dengan gambar
        img = cv2.imread(path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Nama: {names[i]}, Usia (perkiraan): {age} tahun")
        plt.show()

    except Exception as e:
        print(f"Gagal menganalisis {path}: {e}")

# Menampilkan hasil akhir
print("\nüìä HASIL DETEKSI WAJAH DAN USIA:")
for r in results:
    print(f"Nama: {r[0]:10s} | Usia (perkiraan): {r[1]} tahun")
```

```python
# Install library
!pip install opencv-python matplotlib

# Import library
import cv2
import matplotlib.pyplot as plt

# Upload gambar (3 foto wajah)
from google.colab import files
uploaded = files.upload()  # upload orang1.jpg, orang2.jpg, orang3.jpg

# Data nama dan usia
data_wajah = [
    {"file": "pipit.jpeg", "nama": "Pipit", "usia": 20}
]

# Load classifier deteksi wajah bawaan OpenCV
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Loop untuk menampilkan hasil deteksi
for data in data_wajah:
    # Baca gambar
    img = cv2.imread(data["file"])
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Deteksi wajah
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    # Gambar kotak dan teks nama+usia
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(img, f"{data['nama']}, {data['usia']} th", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (36,255,12), 2)

    # Tampilkan hasil
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f"Wajah: {data['nama']} (Usia: {data['usia']} th)")
    plt.show()
```

```python
# ====================================================
# Instalasi dan Import Library
# ====================================================
!pip install opencv-python matplotlib

import cv2
import matplotlib.pyplot as plt
from google.colab import drive
import os

# ====================================================
# 1. Hubungkan Google Drive
# ====================================================
drive.mount('/content/drive')

# Ganti path_folder di bawah ini sesuai lokasi folder kamu di Drive
# (pastikan folder berisi semua foto wajah)
path_folder = "/content/drive/MyDrive/Semester 7/Pengolahan Citra/Praktikum/data wajah"

# ====================================================
# 2. Siapkan data wajah (tanpa CSV, langsung dalam kode)
# ====================================================
data_wajah = [
    {"file": "pipit.jpeg", "nama": "Pipit", "usia": 20},
    {"file": "fila.jpg", "nama": "fila", "usia": 22},
    {"file": "inul.jpeg", "nama": "inul", "usia": 21},
    {"file": "inul.png", "nama": "inul", "usia": 19},
    {"file": "ipul.jpeg", "nama": "ipul", "usia": 20},
    {"file": "nisfa.jpg", "nama": "nisfa", "usia": 23},
    {"file": "lala.jpg", "nama": "lala", "usia": 22},
    {"file": "adel.jpg", "nama": "adel", "usia": 24},
    {"file": "roki.jpeg", "nama": "roki", "usia": 25},
    {"file": "zahra.jpeg", "nama": "zahra", "usia": 21}
]

print("Data wajah ditemukan:")
for d in data_wajah:
    print(f"- {d['file']} ({d['nama']}, {d['usia']} th)")

# ====================================================
# 3. Load classifier wajah bawaan OpenCV
# ====================================================
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# ====================================================
# 4. Loop deteksi wajah per gambar
# ====================================================
for data in data_wajah:
    file_path = os.path.join(path_folder, data["file"])
    nama = data["nama"]
    usia = data["usia"]

    # Pastikan file gambar ada
    if not os.path.exists(file_path):
        print(f"‚ö†Ô∏è File tidak ditemukan: {file_path}")
        continue

    # Baca gambar dan ubah ke grayscale
    img = cv2.imread(file_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Deteksi wajah
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    # Gambar kotak dan tulis nama + usia
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(img, f"{nama}, {usia} th", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (36,255,12), 2)

    # Tampilkan hasil
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f"Wajah: {nama} (Usia: {usia} th)")
    plt.show()
```
